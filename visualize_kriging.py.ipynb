```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
input_file = 'C:\\Users\\Sun\\source\\repos\\ImageProcessing\\Point_IN.txt'
input_data = pd.read_csv(input_file)

# 
output_file = 'C:\\Users\\Sun\\source\\repos\\ImageProcessing\\Point_OUT.txt'
output_data = pd.read_csv(output_file)

# 
plt.figure(figsize=(10, 6))
plt.scatter(input_data['X'], input_data['Y'], c=input_data['Value'], cmap='viridis', marker='o')
plt.colorbar(label='Value')
plt.title('Original Points Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

# 
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(output_data['X'], output_data['Y'], output_data['Value'], c=output_data['Value'], cmap='viridis')
plt.colorbar(sc, label='Interpolated Value')
ax.set_title('Kriging Interpolated Surface')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Value')
plt.show()
```
